```python
import torch
from torch import nn
from torch.autograd import Variable
from torchviz import make_dot, make_dot_from_trace
```

```python
class WAZIR_BLOCK1(nn.Module):
    def __init__(self):
        super().__init__()
        self.CONVOLUTION_LAYER_1 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_1 = nn.PReLU()
        self.drop1 = nn.Dropout2d()
        self.CONVOLUTION_LAYER_2 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_2 = nn.PReLU()
        self.drop2 = nn.Dropout2d()
        self.CONVOLUTION_LAYER_3 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_3 = nn.PReLU()
        self.drop3 = nn.Dropout2d()
    def forward(self,x):
        C1 = self.CONVOLUTION_LAYER_1(x)
        P1 = self.PRELU_1(C1)
        D1 = self.drop1(P1)
        C2 = self.CONVOLUTION_LAYER_1(x)
        P2 = self.PRELU_2(C2)
        D2 = self.drop2(P2)
        C3 = self.CONVOLUTION_LAYER_1(x)
        P3 = self.PRELU_3(C3)
        D3 = self.drop3(P3)
        out = torch.cat((D1,D2,D3),1)
        return out
model1 = WAZIR_BLOCK1()
print(model1)
```

```python
x = Variable(torch.randn(1,1,33,33))
y = model1(x)
make_dot(y.mean(),params=dict(model1.named_parameters()))
```

```python
from prettytable import PrettyTable

def count_parameters(model1):
    table = PrettyTable(["Modules", "Parameters"])
    total_params = 0
    for name, parameter in model1.named_parameters():
        if not parameter.requires_grad: continue
        param = parameter.numel()
        table.add_row([name, param])
        total_params+=param
    print(table)
    print(f"Total Trainable Params: {total_params}")
    return total_params
    
count_parameters(model1)
```

```python
import torch
from torch import nn
from torch.autograd import Variable
from torchviz import make_dot, make_dot_from_trace
```

```python
class WAZIR_BLOCK1(nn.Module):
    def __init__(self):
        super().__init__()
        self.CONVOLUTION_LAYER_1 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_1 = nn.PReLU()
        self.conv2_drop = nn.Dropout2d()
        self.CONVOLUTION_LAYER_2 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_2 = nn.PReLU()
        self.conv2_drop = nn.Dropout2d()
        self.CONVOLUTION_LAYER_3 = nn.Conv2d(1,96,8, stride=16,bias=False)
        self.PRELU_3 = nn.PReLU()
        self.conv2_drop = nn.Dropout2d()
    def forward(self,x):
        C1 = self.CONVOLUTION_LAYER_1(x)
        P1 = self.PRELU_1(C1)
        D1 = self.conv2_drop(P1)
        C2 = self.CONVOLUTION_LAYER_2(x)
        P2 = self.PRELU_2(C2)
        D2 = self.conv2_drop(P2)
        C3 = self.CONVOLUTION_LAYER_3(x)
        P3 = self.PRELU_3(C3)
        D3 = self.conv2_drop(P3)
        out1 = torch.cat((C1,P1,D1,C2,P2,D2,C3,P3,D3),1)
        return out1
model1 = WAZIR_BLOCK1()
print(model1)
```
